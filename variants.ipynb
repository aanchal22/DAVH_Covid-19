```python
import pandas as pd
import numpy as np
import datetime
```

```python
folderpath = "data/us_covid19_vaccine_cases_deaths_daily_count.csv"
df = pd.read_csv(folderpath)
df.info()
```

```python
df['date'] = pd.to_datetime(df['date']).dt.date
df
```

```python
df.loc[:, 'Variant'] = np.where(df['date'] <= datetime.datetime(2020, 9, 30).date(), 'Alpha', np.where((df['date'] > datetime.datetime(2020, 9, 30).date()) & (df['date'] <= datetime.datetime(2021, 7, 31).date()), 'Delta', 'Omicron'))
```

```python
df
```

```python
df.to_csv('data/us_covid19_vaccine_cases_deaths_daily_count_variant.csv', index=False)
```
