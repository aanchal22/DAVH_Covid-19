```python
import pandas as pd
import glob
```

```python
folderpath = "..\COVID-19\csse_covid_19_data\csse_covid_19_daily_reports_us\*.csv"
files = []
for file in glob.glob(folderpath):
    files.append(file)
```

```python
final_df = pd.DataFrame()
for file in files:
    df = pd.read_csv(file)
    df.loc[:, "date"] = file.split('\\')[4].replace('.csv', '')
    final_df = final_df.append(df)
    print(len(df), len(final_df), file.split('\\')[4])
```

```python
final_df.to_csv('cases_deaths_covid-19.csv', index=False)
```

```python
final_df['date'] = pd.to_datetime(final_df['date'], format='%m-%d-%Y')
final_df
```

```python
vaccine_df = pd.read_csv('data/people_vaccinated_us_timeline.csv')
vaccine_df['Date'] = pd.to_datetime(vaccine_df['Date'], format='%Y-%m-%d')
vaccine_df
```

```python
merged_df = final_df.merge(vaccine_df, left_on=['Province_State', 'date'], right_on=['Province_State', 'Date'], how='left')
merged_df
```
